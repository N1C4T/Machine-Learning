```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    coef = bias = 0
    iterations = 3000
    n = len(x)
    learning_rate = 0.06
    plt.scatter(x, y, color ='red', marker = '+')
    for i in range(iterations):
        y_pred = coef*x + bias
        loss = (1/n)*(np.sum([val**2 for val in (y - y_pred)]))
        der_coef = -(2/n)*(np.sum(x*(y - y_pred)))
        der_bias = -(2/n)*(np.sum(y - y_pred))
        coef = coef - learning_rate*der_coef
        bias = bias - learning_rate*der_bias

        plt.plot(x, y_pred, color = 'green')
        print(f'coefficient: {coef}, bias: {bias}, loss(error): {loss}, epoch: {i}')
    
        
    
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
